#### Ejemplos de ciclo while

```python
#Este programa solicita al usuario un número final y realiza un conteo ascendente desde 1 hasta ese valor

limite = int(input("¿Hasta qué número quieres contar? ")) # Solicita un límite y lo convierte a entero [2, 3]
i = 1 # Inicializa la variable de control en 1 [1]
while i <= limite: # Repite el bloque mientras i sea menor o igual al límite ingresado [1, 4]
    print(i) # Muestra en pantalla el valor actual del contador [1, 5]
    i += 1 # Incrementa el contador en 1 para la siguiente vuelta del ciclo [1, 6]
```

```python
# El ciclo procesa números ingresados por el usuario y los suma hasta que se introduce un número negativo, lo cual detiene la ejecución (valor centinela).
suma = 0 # Variable para acumular el total de los números ingresados [7, 9]
while True: # Crea un bucle infinito que se controla internamente [7]
    numero = int(input("Ingresa un número (negativo para salir): ")) # Pide un dato al usuario [7, 10]
    if numero < 0: # Evalúa si el número ingresado cumple la condición de salida [7, 11]
        break # Termina el ciclo de forma inmediata si el número es negativo [7, 12]
    suma += numero # Suma el valor ingresado al acumulador total [7, 13]
print(f"La suma total es: {suma}") # Muestra el resultado acumulado final al usuario [7, 14]
```

```python
#Este ejemplo calcula el producto acumulado de un número dado por el usuario en orden descendente hasta llegar a uno.
n = int(input("Ingresa un número para su factorial: ")) # Obtiene el número base desde el teclado [2, 7]
f = 1 # Inicializa el acumulador del producto en 1 [7, 15]
original = n # Guarda el valor original para mostrarlo en el mensaje final [16]
while n > 0: # El ciclo se ejecuta mientras el número sea mayor que cero [7, 17]
    f *= n # Multiplica el acumulador por el valor actual de n [7, 13]
    n -= 1 # Reduce el valor de n en una unidad por cada iteración [18, 19]
print(f"El factorial de {original} es {f}") # Informa el resultado final del cálculo [7, 15]
```

```python
# El programa obliga al usuario a ingresar un número de mes válido (entre 1 y 12). Si el dato es incorrecto, el ciclo se repite hasta obtener una entrada válida.
mes = 0 # Inicializa la variable con un valor que fuerza la entrada al ciclo [20]
while mes < 1 or mes > 12: # El ciclo continúa mientras el valor esté fuera del rango 1-12 [20, 21]
    mes = int(input("Introduce un número de mes (1-12): ")) # Solicita el dato al usuario [10, 20]
    if mes < 1 or mes > 12: # Verifica si la entrada sigue siendo inválida [20, 22]
        print("Valor inválido. Intenta de nuevo.") # Imprime un aviso de error para el usuario [5, 20]
print(f"Has seleccionado un mes válido: {mes}") # Muestra el valor correcto una vez aceptado
```

```python
# Utiliza un ciclo para repetir una pregunta hasta que el usuario logre coincidir su entrada con un número previamente definido en el código.
secreto = 7 # Define el número que el usuario debe descubrir [23]
intento = 0 # Inicializa la variable de respuesta del usuario [24]
while intento != secreto: # Repite el proceso mientras el intento sea diferente al secreto [25]
    intento = int(input("Adivina el número secreto: ")) # Solicita la apuesta del usuario [2, 23]
    if intento != secreto: # Compara si el intento fue fallido [26]
        print("Incorrecto, ¡sigue intentando!") # Da retroalimentación negativa al usuario [5, 23]
print("¡Felicidades! Adivinaste el número.") # Envía un mensaje de éxito al salir del ciclo [23]
```
